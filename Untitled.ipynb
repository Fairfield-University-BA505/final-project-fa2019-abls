```python
# Replace missing values with the average for the column because they are continuous values
wine_data['price'] = wine_data['price'].fillna(wine_data['price'].mean())
# Check to see if all values are filled
wine_data.isnull().mean()*100
```

```python
scatter_plot = plt.figure(figsize=(9,8))
axes2 = scatter_plot.add_subplot(1,1,1)
axes2.scatter(wine_data.points, wine_data.price, s=8) # Sets points as the x-axis and price as the y-axis
axes2.set_title('Price per bottle vs. Points')
axes2.set_ylabel('Price per bottle')
axes2.set_xlabel('Points')
scatter_plot.show()
```

```python
# Calculate average price and score per bottle of wine by country
means = wine_data.groupby('country').mean()
# Create columns to determine average value by country 
means['value'] = means['points'] / means['price']
means
```

```python
# Create a bar chart to display average price by country
wine_data['price'].groupby(wine_data['country']).mean().plot(kind='bar', figsize=(10,9))
plt.title('Average Price by Country')
plt.ylabel('Average Price - $')
```

```python
# Create a bar chart to display average price by country
wine_data['points'].groupby(wine_data['country']).mean().plot(kind='bar', figsize=(10,9))
plt.title('Average Points by Country')
plt.ylabel('Points')
ax = plt.gca()
ax.set_ylim([80,100])
```

```python
# Create a bar chart to display average value by country
means.plot(y = 'value', kind='bar', figsize=(10,9), legend=False)
plt.title='Average Value by Country'
plt.ylabel='Value'
plt.show()
```
